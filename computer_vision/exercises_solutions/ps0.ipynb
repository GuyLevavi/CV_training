```python
green_baboon_img = baboon_img[..., 1]
blue_peppers_img = peppers_img[..., 0]
# Get image dimensions
h, w = green_baboon_img.shape[:2]

# Define the patch size
patch_size = 100
half_patch = patch_size // 2

# Compute the center
center_y = h // 2
center_x = w // 2

yslice = slice(center_y - half_patch, center_y + half_patch)
xslice = slice(center_x - half_patch, center_x + half_patch)

# Extract the patch
blue_peppers_img[yslice, xslice] = green_baboon_img[yslice, xslice]
plt.imshow(blue_peppers_img, cmap='gray')
```
